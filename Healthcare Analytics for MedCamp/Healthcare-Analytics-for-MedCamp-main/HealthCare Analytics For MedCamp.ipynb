```python
# Importing necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split # Used to split the dataset into training and testing sets
from sklearn.linear_model import LinearRegression #  Used to create a linear regression model
import os  # Add this line to import the os module
from IPython.display import display

# Setting the directory where the dataset is located
dataset_directory = 'dbs_dataset'
# Getting the files from the directory and printing the path of each file
for dirname, _, filenames in os.walk(dataset_directory):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
#Preparation and Cleaning of data
# Reading the data dictionary from an Excel file
dict_data = pd.read_excel("dbs_dataset/dbs_Train_dataset/Data_Dictionary.xlsx")
# Reading data related to the first health camp attendance
first_camp_data = pd.read_csv("dbs_dataset/dbs_Train_dataset/First_Health_Camp_Attended.csv")
# Reading data related to the second health camp attendance
second_camp_data = pd.read_csv("dbs_dataset/dbs_Train_dataset/Second_Health_Camp_Attended.csv")
# Reading data related to the third health camp attendance
third_camp_data = pd.read_csv("dbs_dataset/dbs_Train_dataset/Third_Health_Camp_Attended.csv")
# Reading health camp details data
camp_details_data = pd.read_csv("dbs_dataset/dbs_Train_dataset/Health_Camp_Detail.csv")
# Reading patient profiles data
patient_profiles_data = pd.read_csv("dbs_dataset/dbs_Train_dataset/Patient_Profile.csv")
# Reading test data
testing_data= pd.read_csv("dbs_dataset/dbs_Train_dataset/test.csv")
# Reading training data
training_data= pd.read_csv("dbs_dataset/dbs_Train_dataset/Train.csv")
```

```python
# Merging patient profiles with the first health camp data on "Patient_ID"
merged_data_one = pd.merge(right=patient_profiles_data, left=first_camp_data, on="Patient_ID")

# Displaying information about the merged dataset
print("\n\033[1mInformation about the merged dataset\033[0m\n")
display(merged_data_one.info())

# Merging the result with health camp details on "Health_Camp_ID"
merged_data_two = merged_data_one.merge(camp_details_data, on="Health_Camp_ID")

# Displaying information about the updated merged dataset
print("\n\033[1mInformation about the updated merged dataset\033[0m\n")
display(merged_data_two.info())

# Merging the updated dataset with the second health camp data on "Patient_ID"
merged_data_three = merged_data_two.merge(second_camp_data, on="Patient_ID")

# Displaying information about the further updated merged dataset
print("\n\033[1mInformation about the updated merged dataset\033[0m\n")
display(merged_data_three.info())

# Merging the dataset once more with the third health camp data on "Patient_ID"
dataset_merged = merged_data_three.merge(third_camp_data, on="Patient_ID")

# Displaying the first few rows of the final merged dataframe
print("\n\033[1mfirst few rows of the final merged dataframe\033[0m\n")
display(dataset_merged.head())
```

```python

# Apply machine learning to make predictions
dumm_merged = pd.get_dummies(dataset_merged, prefix=None, prefix_sep="_", drop_first=False)
X = dumm_merged.iloc[:, :-1]
y = dumm_merged.health
display(X.shape)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=23)

model = LinearRegression()
model.fit(X_train, y_train)

# Evaluate the model's performance on the test set
rsq_train = model.score(X_train, y_train)
rsq_test = model.score(X_test, y_test)
print("R-squared (Training):", rsq_train)
print("R-squared (Testing):", rsq_test)
```
